# papularity base recommendation system

#### we take top fifty books on the basis of rating but we only take those culmns there rating is grater then 250

```python
#mearge tow data frames
rattings_with_name=rattings.merge(book,on='ISBN')
```

```python
rattings_with_name.sample(5)
```

```python
num_ratting_df=rattings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_ratting_df.rename(columns={'Book-Rating':'num-rating'},inplace=True)
num_ratting_df
```

```python
avg_ratting_df=rattings_with_name.groupby('Book-Title').mean()['Book-Rating'].reset_index()
avg_ratting_df.rename(columns={'Book-Rating':'avg-rating'},inplace=True)
avg_ratting_df
```

```python
papular_df=num_ratting_df.merge(avg_ratting_df,on='Book-Title')
```

```python
papular_df
```

```python
papular_df=papular_df[papular_df['num-rating']>250].sort_values('avg-rating',ascending=False).head(50)
```

```python
papular_df=papular_df.merge(book,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','num-rating','avg-rating']]
```

# Collaborative Filtring based Recommender System

#### we considred those users who minimum rates on 200 books && we considred those books on which minimum 50 person rates

```python
x=rattings_with_name.groupby('User-ID').count()['Book-Rating'] > 200
ratting_krny_valy_users=x[x].index
```

```python
filtered_ratting=rattings_with_name[rattings_with_name['User-ID'].isin(ratting_krny_valy_users)]
```

```python
y=filtered_ratting.groupby('Book-Title').count()['Book-Rating'] >= 50
famous_book=y[y].index

final_ratting=filtered_ratting[filtered_ratting['Book-Title'].isin(famous_book)]
pt = final_ratting.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')

pt.fillna(0,inplace=True)
pt
```

```python
papular_df['Image-URL-M'][0]
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score=cosine_similarity(pt)
similarity_score.shape
```

```python
def recommend(book_name):
    #first we fetch index with book name
    index = np.where(pt.index==book_name)[0][0]
    #now find similar items
    #enumerate means show index with similarity score each book
    similar_items=sorted(list(enumerate(similarity_score[index])),key=lambda x:x[1],reverse=True)[1:6]
    data=[]
    for i in similar_items:
        item=[]
        temp_df=book[book['Book-Title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        data.append(item)
    return data
```

```python
recommend('The Five People You Meet in Heaven')
```
